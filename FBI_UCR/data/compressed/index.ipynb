```python
import os
```

```python
files = [file for file in os.listdir(".") if file[-3:] == "zip"]

state_data = {}

for file in files:
    state = file.split("-")[0]
    year = file.split("-")[1][:4]
    if state not in state_data.keys():
        state_data[state] = [year]
    else:
        state_data[state].append(year)
        state_data[state].sort()
state_data
```
